##Outliers

```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
#Dataset:
dataset = [11,10,12,14,12,15,102,12,14,17,19,107,10,12,108, 12,11,14,13,15,12,10,14,13,15,10]
```

```python
plt.hist(dataset)
```

```python
#Z-Score:
outliers=[]
def detect_outliers(data):
  threshold = 3 ##rd Standard Devaition
  mean = np.mean(data)
  std = np.mean(data)

  for i in data:
    z_score = (i-mean)/std
    if np.abs(z_score) > threshold:
      outliers.append(i)
  return outliers
```

```python
detect_outliers(dataset)
```

##Measure of Dispersion:

```python
statistics.variance(ages) #Sample Variance
```

```python
np.var(ages, axis=0) #Population Variance
```

```python
#Population Variance:
def variance(data):
  n = len(ages)
  mean = sum(data)/n
  deviation = [(x-mean)**2 for x in data]
  variance = sum(deviation)/n
  return variance
```

```python
variance(ages)
```

```python
#Sample Variance:
def variance(data):
  n = len(ages)
  mean = sum(data)/n
  deviation = [(x-mean)**2 for x in data]
  variance = sum(deviation)/(n-1)
  return variance
```

```python
variance(ages)
```

```python
statistics.pvariance(ages)
```

```python
import math
math.sqrt(statistics.pvariance(ages))
```

##Check whether Distribution is Normal or Not:

```python
##QQ Plot
import matplotlib.pyplot as plt
import scipy.stats as stat
import pylab
```

```python
def plot_data(sample):
  plt.figure(figsize=(10,6))
  plt.subplot(1,2,1)
  sns.histplot(sample)
  plt.subplot(1,2,2)
  stat.probplot(sample, dist='norm',plot=pylab)
  plt.show()
```

```python
s = np.random.normal(0.5, 0.2, 1000)
plot_data(s)
```

```python
plot_data(df['sepal_width'])
```

```python
mu, sigma = 3,1
sample = np.random.lognormal(mu, sigma, 100)
```

```python
def plot_data_log(sample):
  plt.figure(figsize=(10,6))
  plt.subplot(1,2,1)
  sns.histplot(sample)
  plt.subplot(1,2,2)
  stat.probplot(np.log(sample), dist='norm', plot=pylab)
  plt.show()
```

```python
plot_data_log(sample)
```
